El presente proyecto tiene como objetivo crear una aplicación interactiva de consola que permita al usuario calcular propiedades esenciales de las figuras circulares. Es importante distinguir las fórmulas matemáticas que sustentan este programa:

1. Área del Círculo: Representa la superficie contenida dentro de la circunferencia. La fórmula es A=π⋅r2A=π⋅r 2 , donde r es el radio.
2. Perímetro (Circunferencia): Es la longitud del borde del círculo. Se calcula como P=2⋅π⋅rP=2⋅π⋅r.
3. Volumen de la Esfera: Aunque el menú menciona "círculo", el volumen es una propiedad tridimensional, por lo que corresponde a una esfera. La fórmula es V=43⋅π⋅r3V= 34​ ⋅π⋅r 3 .

```python
# Elaborado por ✨Karol Castillo
import math

# Función para calcular el área del círculo
def areacirculo():
    print("\n🔵 --- Cálculo del Área del Círculo ---")
    radio = float(input("Por favor, ingresa el radio del círculo: "))
    # Fórmula: A = pi * r^2
    area = math.pi * (radio ** 2)
    print(f"✨ El área del círculo con radio {radio} es: {area:.2f}")

# Función para calcular el perímetro del círculo
def perimetrocirculo():
    print("\n🟢 --- Cálculo del Perímetro (Circunferencia) ---")
    radio = float(input("Por favor, ingresa el radio del círculo: "))
    # Fórmula: P = 2 * pi * r
    perimetro = 2 * math.pi * radio
    print(f"✨ El perímetro del círculo con radio {radio} es: {perimetro:.2f}")

# Función para calcular el volumen de la esfera
def volumencirculo():
    print("\n🟣 --- Cálculo del Volumen de la Esfera ---")
    radio = float(input("Por favor, ingresa el radio de la esfera: "))
    # Fórmula: V = (4/3) * pi * r^3
    volumen = (4/3) * math.pi * (radio ** 3)
    print(f"✨ El volumen de la esfera con radio {radio} es: {volumen:.2f}")

# Función para mostrar el menú y controlar el flujo
def menu():
    opcion = 0
    while opcion != 4:
        print("\n" + "="*40)
        print("🌟 MENÚ GEOMÉTRICO INTERACTIVO 🌟")
        print("="*40)
        print("1️⃣  Calcular Área del Círculo")
        print("2️⃣  Calcular Perímetro del Círculo")
        print("3️⃣  Calcular Volumen de la Esfera")
        print("4️⃣  Salir del Programa")
        print("="*40)

        # Se asume entrada válida según instrucciones (sin validación extra)
        opcion = int(input("👉 Selecciona una opción (1-4): "))

        if opcion == 1:
            areacirculo()
        elif opcion == 2:
            perimetrocirculo()
        elif opcion == 3:
            volumencirculo()
        elif opcion == 4:
            print("\n👋 ¡Gracias por usar el programa! Hasta luego. 🚀")
        else:
            print("\n⚠️  Opción no válida. Por favor intenta de nuevo.")

# Llamada inicial a la función principal
if __name__ == "__main__":
    menu()
```
